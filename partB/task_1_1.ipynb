## Step-by-Step Method Description

*   **Step 1:** Explicit Data Mapping
    *   **Description:** Instead of implicitly mapping data to a high-dimensional space using the kernel trick (which is evaluated via inner products $K(x_i, x_j)$), the proposed method explicitly transforms the input feature vector $x$ into a high-dimensional mapped vector $\phi(x)$ using a low-degree polynomial mapping. For instance, with a degree $d=2$ and neglecting the scaling coefficient $\sqrt{2}$ for simplicity, the vector explicitly becomes $\phi(x) = [1, x_1, \dots, x_n, x_1^2, \dots, x_n^2, x_1 x_2, \dots, x_{n-1} x_n]^T$.
    *   **Reference:** Section 3.1, Equation (5).
    *   **Purpose:** To explicitly represent the data in a higher-dimensional space where it is more likely to be linearly separable, allowing the use of fast linear classification algorithms.

*   **Step 2:** Training via Fast Linear SVM Solvers
    *   **Description:** The algorithm trains a linear Support Vector Machine directly on the explicitly mapped data $\phi(x)$ using decomposition methods (specifically coordinate descent), rather than solving the dual problem with an $n \times n$ dense kernel matrix $Q$. This changes the computational cost of a decomposition iteration from $O(l \bar{n})$ (where $l$ is the number of instances and $\bar{n}$ is the average number of non-zero features per instance) down to $O(\hat{n})$ (where $\hat{n}$ is the average number of non-zeros in the mapped vector $\phi(x)$).
    *   **Reference:** Section 3.2, Equation (8).
    *   **Purpose:** To significantly speed up the training process by taking advantage of the sparsity in the explicitly mapped vectors, bypassing the slow kernel evaluation steps required in traditional non-linear SVMs.

*   **Step 3:** Dynamic Computation and Sparse Representation
    *   **Description:** To handle the massive dimensionality of $\phi(x)$ without running out of memory, the algorithm dynamically calculates $\phi(x_i)$ on the fly during the training loop instead of pre-computing and storing $O(l \cdot \hat{n})$ components. Furthermore, it leverages a sparse implementation for the weight vector $w$ (or standard dense vector combined with hashing if the data is highly sparse like in NLP tasks).
    *   **Reference:** Section 3.4 (Implementation Issues) and Section 4.3 (Table 3).
    *   **Purpose:** To prevent memory exhaustion (out-of-memory errors) and minimize L2 cache misses, thereby ensuring the overall training speed remains extremely fast despite the combinatorial explosion of polynomial features.

## Final Summary Sentence
This paper solves the problem of prohibitively slow training and testing times in traditional nonlinear kernel SVMs on large-scale datasets, and the authors claim their explicit low-degree polynomial mapping approach combined with linear SVM solvers is significantly faster than standard kernel methods while achieving comparable classification accuracy to highly nonlinear alternatives like the RBF kernel.
