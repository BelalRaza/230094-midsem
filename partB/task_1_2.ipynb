# Task 1.2: Key Assumptions

## Assumption 1
*   **Assumption:** The dataset exhibits sufficient feature sparsity such that the average number of non-zero elements in the mapped polynomial vector ($\hat{n}$) is significantly smaller than the cost of evaluating the kernel across all instances ($l \bar{n}$).
*   **Why the method needs it:** The primary speedup mechanism of explicitly formulating the linear SVM on $\phi(x)$ depends on the cost of a decomposition iteration being $O(\hat{n})$ rather than $O(l \bar{n})$. If $\hat{n}$ is larger than $l \bar{n}$, calculating the explicit mapping becomes more computationally expensive than traditional kernel approaches.
*   **Violation scenario:** A dense dataset with a large number of continuously valued features where almost no values are strictly zero (e.g., raw, uncompressed high-resolution images or dense physiological signal recordings). In this strictly dense case, an explicit polynomial expansion of degree $d$ scales as $O(n^d)$, rapidly overwhelming the computational savings.
*   **Paper reference:** Section 3.2 ("If $\hat{n} < l \bar{n}$, the new strategy may be faster than the training using kernels") and Table 2's comparative analysis.

## Assumption 2
*   **Assumption:** Data is structurally simple enough that a low-degree polynomial mapping (specifically $d=2$ or $d=3$) provides sufficient feature interaction and non-linearity to linearly separate the classes with high accuracy.
*   **Why the method needs it:** Due to the combinatorial explosion of terms in the explicit mapping, it is practically intractable to use degrees higher than $d=3$ because generating and storing the combination of all features $x_1^{k_1} x_2^{k_2} \dots$ would crash the memory or cause immense cache slowdowns. Thus, the method relies on a small $d$ being "good enough."
*   **Violation scenario:** Highly complex, non-linear classification problems that require infinite-dimensional transformations to be cleanly separated, such as highly intertwined spiral datasets or complex geographic topological data. In these datasets, low-degree polynomials will fail to capture the required decision boundaries, yielding low accuracy.
*   **Paper reference:** Section 3.1, Section 4.4 (Table 5 illustrates this assumption failing on the `covtype` dataset, where degree-2 polynomial mapping performs identically to standard linear SVM and poorly compared to RBF).

## Assumption 3
*   **Assumption:** Individual features are either highly independent, or follow established structural groupings, such that the expected number of combinations ($x_r x_s \neq 0$) across the entire dataset remains small enough to boundedly fit the non-zero weight vector elements into standard computer memory.
*   **Why the method needs it:** The weight vector $w$ has dimensionality corresponding to all polynomial combinations of features ($O(n^d)$). The system requires that many of these combinations evaluating to non-zero never actually occur in the dataset simultaneously, so that the sparse weight vector $\tilde{n}$ does not explode entirely.
*   **Violation scenario:** A dataset with a massive number of features ($n$) where feature co-occurrences are highly overlapping and fully distributed without distinct structural independence, causing every mathematical combination $x_r x_s$ to appear at least once across instances.
*   **Paper reference:** Section 3.4 Equation (11) (Probability estimation for non-zeros in $w$) and Section 4.2 (The authors had to exclude the `news20` dataset from some experiments because the active $w$ vector exhausted the 16 GB machine memory).
