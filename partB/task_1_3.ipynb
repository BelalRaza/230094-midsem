# Task 1.3: What the Paper Claims to Improve

*   **Baseline or Prior Method:**
    The paper directly compares against traditional non-linear Support Vector Machines utilizing the kernel trick, specifically focusing on the **RBF (Gaussian) kernel implementation built into LIBSVM**. Another baseline referenced is the standard **Linear SVM (via LIBLINEAR)** without any mapping.

*   **Limitation Identified:**
    The authors observe that for massive, large-scale datasets, non-linear SVMs trained via the traditional dual-problem decomposition (like LIBSVM) are exceedingly slow in both training and testing. This is because every iteration required an $O(l \bar{n})$ calculation during training to evaluate the dense kernel, and $O(\#SV \cdot \bar{n})$ during testing to compare against the support vectors.

*   **How the Proposed Method Overcomes It:**
    The proposed method resolves this limitation by mapping the input vector explicitly to a low-degree polynomial space $\phi(x)$ and treating it as a standard linear SVM problem, utilizing fast large-scale linear solvers (like LIBLINEAR). By converting the computational complexity over to $O(\hat{n})$ (which is purely dependent on the non-zero terms of the newly explicitly mapped data), they dramatically bypass the massive sequence of computations involved in iterating the dual formulations via the kernel trick matrix.

*   **Condition Where It Would Not Outperform the Baseline:**
    Based on the experimental observations, the proposed method distinctly fails to outperform the RBF baseline when the actual mathematical distribution of the problem cannot be represented fully within a degree-2 or degree-3 polynomial space. An example condition is when the features exhibit extremely complex conditional relationships on a continuous scale rather than discrete or loosely linked binary relationships. In such environments, the limited polynomial space will "underfit" the problem—resulting in an accuracy strictly below what an RBF kernel (which projects to infinite dimensions) could trivially separate. Additionally, if the dataset consists of $100\%$ dense features (not mostly zeros), explicitly computing all polynomial combinations takes exponentially more time and memory than relying on the kernel trick's simplified internal dot product, causing the explicit mapping to be worse in both speed and memory.
